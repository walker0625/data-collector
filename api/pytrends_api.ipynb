```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
from matplotlib import rc # mac korean font

keyword1 = "땡겨요 서울페이"
period = "today 5-y"

trend_obj = TrendReq()
trend_obj.build_payload(kw_list=[keyword1], timeframe=period, geo="KR")
trend_df = trend_obj.interest_over_time()

# mac korean font
rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

# plt.rc('font', family='NanumGothic') window korean font
plt.style.use("ggplot")
plt.figure(figsize=(14,5))

trend_df[keyword1].plot()

plt.title("땡겨요 서울페이 지난 5년간 검색 트렌드 수", size=15)
plt.legend(loc="best")
plt.show()
```
